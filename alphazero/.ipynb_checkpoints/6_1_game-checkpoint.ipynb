- 듀얼 네트워크
  - 뉴럴 네트워크의 구성을 정의하고 베스트 플레이어 모델을 생성함
  - 현재 국면에 대해, 정책, 가치의 두가지를 출력
  - 최초 가중치는 랜덤으로 부여하는 미학습 상태이므로, 매우 약하다
    - 학습과 평가후 즉시 갱신됨
- 셀프 플레이 파트
  - 베스트 플레이어 모델을 사용하여 셀프 플레이를 수행
  - 일공지능끼리 게임이 종료딜때까지 플레이 하는것
  - 뉴럴 네트워크 학습에 이용할 학습 데이터(정책과 가치)를 생성
- 파라미터 갱신 파트
  - 셀프 플레이에서 생성한 학습 데이터를 사용하여 최신 플레이어를 학습시킨다
  - 최신 플레이어의 초기 상태는 베스트 플레이러를 복사하여 사용
- 신규 파라미터 평가 파트
  - 최신 플레이어와 베스트 플레이어를 대전시킴
  - 충분히 앞선 경우에는 교대
  - 학습을 통해 강해지지 않으면 그 학습을 채택하지 않음 : 알파고, 알파고제로에만 존재하는 방법론임
